```python
# rolling_method와 온도 간 차이?
sns.boxplot(x='rolling_method',y='rolling_temp',data=df_raw)
```

- CR의 평균적으로 압연 공정 온도가 더 높다.
- CR : 제어압연법은 압하비, 압연온도를 제어하여 상온의 조직을 미세화 하여 인성 및 가공성을 개선시킬 수 있는 압연법으로 열처리공정중 노멀라이징(Normalizing)을 대체할 수 있다.
    - 압연 시작 온도 : 1100~1200
    - 최종 압연 온도 : 750~900
- TMCP : 압연과정 직후의 강판을 온라인상에서 급속 냉각하여 급속조직을 미세하게 제어함으로써 강의 인성과 강도를 향상시키는 기술을 적용
    - 압연 시작 온도 : 1000~1100
    - 최종 압연 온도 : 700~900

```python
df_raw = df_raw[df_raw['rolling_temp'] > 600] # rolling_temp 이상치 제거
```

```python
# Filter the dataset for the necessary columns
df_filtered = df_raw[['fur_heat_temp', 'rolling_temp', 'scale']]

# Filter out rows where rolling_temp is 0
df_filtered_no_zero = df_filtered[df_filtered['rolling_temp'] > 0]

# Plotting histograms again for rolling_temp without 0-degree data

plt.figure(figsize=(14, 7))

# Histogram for Furnace Heating Temperature (unchanged)
plt.subplot(1, 2, 1)
sns.histplot(df_filtered_no_zero, x='fur_heat_temp', hue='scale', multiple='stack', palette='coolwarm')
plt.title('Scale Count vs Furnace Heating Temperature')
plt.xlabel('Furnace Heating Temperature (℃)')
plt.ylabel('Scale Count')

# Histogram for Rolling Temperature (after removing 0-degree data)
plt.subplot(1, 2, 2)
sns.histplot(df_filtered_no_zero, x='rolling_temp', hue='scale', multiple='stack', palette='coolwarm')
plt.title('Scale Count vs Rolling Temperature (0-degree data removed)')
plt.xlabel('Rolling Temperature (℃)')
plt.ylabel('Scale Count')

plt.tight_layout()
plt.show()
```
