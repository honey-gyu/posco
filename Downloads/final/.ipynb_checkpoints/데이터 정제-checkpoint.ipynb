### 변수 제거

```python
drop_cols = ['pg1_impfit','pg1_imp','pg1_voltage_start','pg1_curr_end','pg1_box_col','pg1_box_dan','pg1_temp_avg','pc1_imp','pc1_voltage_start','pc1_curr_end','pc1_box_col','pc1_box_dan' \
 ,'sa1_box_col','sa1_box_row','sa1_box_dan','socv1_ocv','sa2_box_col','sa2_box_row','sa2_box_dan','pg1_tat', 'pc1_tat', 'sa1_tat', 'socv1_tat', 'sa2_tat',\
       'socv2_tat', 'sa3_tat', 'socv3_tat', 'sa4_tat','cells_in_tray']

df_all.drop(columns=drop_cols,axis=1,inplace=True)
```

## 파생변수

```python
df_all['defective'] = df_all['judge'].apply(lambda x: 1 if x == '불량' else 0) # 불량이면 1로 아닌 건 0으로 defective에 저장
defect_rate = df_all.groupby('tray_id').agg(total_cells=('cell_id', 'count'), defective_cells=('defective', 'sum'))
defect_rate['defect_rate'] = (defect_rate['defective_cells'] / defect_rate['total_cells']) * 100
```

```python
defect_rate.sort_values(by='defect_rate',ascending=False)
```

```python
df_all = df_all.merge(defect_rate[['defect_rate']], on='tray_id', how='left')
```
