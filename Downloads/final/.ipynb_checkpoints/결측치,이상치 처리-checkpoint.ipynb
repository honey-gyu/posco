'sa3_box_col', 'sa3_box_row', 'sa3_box_dan' -> tray_id 같으면 같다.

```python
process[['sa3_box_col', 'sa3_box_row', 'sa3_box_dan']] = process.groupby('tray_id')[['sa3_box_col', 'sa3_box_row', 'sa3_box_dan']].fillna(method='ffill')

process[process['tray_id'] == 'TRAY-1339'][['tray_id', 'sa3_box_col', 'sa3_box_row', 'sa3_box_dan']]
```

```python
process[process['tray_id']=='TRAY-1339'][['tray_id','m1_thick', 'm1_voltage', 'm1_res_ac',
       'm1_mv', 'm1_voltage_an']]
```

tray 별로 평균 구해서 채우기?

```python
mean_values = process.groupby('tray_id')[['m1_thick', 'm1_voltage', 'm1_res_ac', 'm1_mv', 'm1_voltage_an']].mean()

for col in ['m1_thick', 'm1_voltage', 'm1_res_ac', 'm1_mv', 'm1_voltage_an']:
    process[col] = process.groupby('tray_id')[col].transform(lambda x: x.fillna(mean_values.loc[x.name, col]))

process[process['tray_id'] == 'TRAY-1339'][['tray_id', 'm1_thick', 'm1_voltage', 'm1_res_ac', 'm1_mv', 'm1_voltage_an']]
```

```python
process[process['tray_id']=='TRAY-1339'][['socv2_ocv','socv3_ocv',
       'ocv2_deltaocv', 'ocv1_deltaocv']]
```

tray 별로 평균 구해서 채우기? 최빈값?

```python
mean_values = process.groupby('tray_id')[['socv2_ocv','socv3_ocv','ocv2_deltaocv', 'ocv1_deltaocv']].mean()

for col in ['socv2_ocv','socv3_ocv','ocv2_deltaocv', 'ocv1_deltaocv']:
    process[col] = process.groupby('tray_id')[col].transform(lambda x: x.fillna(mean_values.loc[x.name, col]))

process[process['tray_id'] == 'TRAY-1339'][['socv2_ocv','socv3_ocv','ocv2_deltaocv', 'ocv1_deltaocv']]
```
