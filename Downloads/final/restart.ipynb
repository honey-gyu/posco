```python
df_all = df_all.drop(columns = ['sa1_box_col', 'sa1_box_row', 'sa1_box_dan', 'socv1_ocv', 'sa2_box_col',
                          'sa2_box_row', 'sa2_box_dan', 'socv2_ocv', 'sa3_box_col', 'sa3_box_row', 
                          'sa3_box_dan', 'socv3_ocv', 'ocv2_deltaocv', 'ocv1_deltaocv', 'm1_thick', 
                          'm1_voltage', 'm1_res_ac', 'm1_mv', 'm1_voltage_an'])
df_all
```

```python
df_all['defective'] = df_all['judge'].apply(lambda x: 1 if x == '불량' else 0) # 불량이면 1로 아닌 건 0으로 defective에 저장

defect_rate = df_all.groupby('tray_id').agg(total_cells=('cell_id', 'count'), defective_cells=('defective', 'sum'))
# 셀_id의 수는 = total_cells
# defective(양품, 불량을 범주화->숫자화)의 합 = defective_cells 

defect_rate['defect_rate'] = (defect_rate['defective_cells'] / defect_rate['total_cells']) * 100
# 백분율
```
