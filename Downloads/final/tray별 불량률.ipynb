```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib
%matplotlib inline

import warnings
warnings.filterwarnings('ignore')
```

```python
process = pd.read_csv('./C2_이차전지/bat_process.csv',encoding='euc-kr')
process.head()
```

트레이별로 셀 개수는 최대 25개, 25개가 안 채워져있는 경우도 존재

```python
process['defective'] = process['judge'].apply(lambda x: 1 if x == '불량' else 0)

defect_rate = process.groupby('tray_id').agg(
    total_cells=('cell_id', 'count'),
    defective_cells=('defective', 'sum')
)

process.drop('defective',axis=1,inplace=True)

defect_rate['defect_rate'] = (defect_rate['defective_cells'] / defect_rate['total_cells']) * 100
```
