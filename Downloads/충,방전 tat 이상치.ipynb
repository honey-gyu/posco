```python
# 숫자형 변수들만 선택
numeric_columns = df.select_dtypes(include=[np.number])

# 모든 숫자형 변수들 간의 상관관계 행렬 계산
correlation_matrix = numeric_columns.corr()

# 상관관계 행렬 시각화
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, cmap='coolwarm', annot=False, fmt=".2f")
plt.title('상관관계 행렬')
plt.show()
```

```python
# 상관계수가 0.8 이하인 값들을 0으로 설정하여 필터링
filtered_corr_matrix = correlation_matrix.copy()
filtered_corr_matrix[filtered_corr_matrix.abs() < 0.8] = 0

# 히트맵에 나온 상관계수 0.8 이상인 변수들만 추출
selected_variables = filtered_corr_matrix.columns[(filtered_corr_matrix != 0).any()].tolist()

# 해당 변수들만 포함한 데이터프레임 생성
filtered_df = df[selected_variables]

# 결과 확인 (필요에 따라 생략 가능)
filtered_df.head()
```

```python
filtered_df.info()
```

```python
# filtered_df의 상관관계 행렬 계산 및 히트맵 시각화
filtered_corr_matrix_df = filtered_df.corr()

plt.figure(figsize=(12, 10))
sns.heatmap(filtered_corr_matrix_df, cmap='coolwarm', annot=False, fmt=".2f")
plt.title('Filtered_df 데이터프레임의 상관관계 히트맵')
plt.show()
```
