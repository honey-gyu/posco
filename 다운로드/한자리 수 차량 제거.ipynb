```python
df['Mileage'] = df['Mileage'].str.replace('kmpl','').astype('float')
df['Engine'] = df['Engine'].str.replace('CC','').astype('float')
df['Power'] = df['Power'].str.replace('bhp','').astype('float')
```

```python
df.info()
```

```python
# Name열 split 후 회사명(브랜드)만 리턴하는 함수
def extract_brand(text):
    return text.split()[0]

df['Brand'] = df['Name'].apply(extract_brand)
df['Brand'].head(10)
```

# 차량(브랜드) 행 중 한자리 수 제거 시작

```python
# 브랜드 함수의 값 수 확인
# 한자리 수 차량 5개 제거 시작
brand_counts = df['Brand'].value_counts()
brand_counts
```

```python
# 브랜드 행을 brand_count로 map하고 행 추가
df['Brand_count'] = df['Brand'].map(brand_counts)
df
```

```python
# brand_count 행에서 10개 미만이 있는지 query 함수로 필터링 해줘
filtered_df = df.query("Brand_count < 10")
print(filtered_df)

# 총 10개 있음을 확인
```

```python
# brand_count 수가 10개 미만인 열을 제거함
index_to_drop = df[df['Brand_count'] < 10].index
df.drop(index_to_drop, inplace=True)
print(df)
```

```python
# 조건 필터링으로 제거된지 확인 / 결과 empty라고 나옴
filtered_df = df.query("Brand_count < 10")
print(filtered_df)
```

# 여기부터는 원본파일_v3이므로 이 전사이에 추가하면 됨

```python
# Power 기준 차량 크기 구분하는 함수
# 경차 -> 1000 미만
# 소형 -> 1600 미만
# 중형 -> 2000 미만
# 대형 -> 2000 이상
def car_size(cc):
    if cc >= 2000:
        car_size = 'full'
    elif cc >= 1600:
        car_size = 'mid'
    elif cc > 1000:
        car_size = 'small'
    else:
        car_size = 'light'
        
    return car_size

df['Size']= df['Engine'].apply(car_size)
```
