```python
# p232 연습
# 정규분포
# → A/S 작업을 수행하는데 걸리는 시간을 측정해 본 결과, 평균 115분, 표준편차 20분이 소요된다는 것을 알았음.
# → 만일 A/S 작업시간의 허용범위가 135분 이내라면 135분 이상 걸리는 A/S 작업비율은 어느 정도일까요?

# 누적확률 값 계산
mu = 115 # 평균
sigma = 20 # 표준편차
x = 135 # 허용범위
prob = stats.norm.cdf(x, mu, sigma)
print("{0}분 이상 걸리는 A/S 작업 비율: {1:.1f}%".format(x, (1-prob)*100))

# 정규분포확률 구하기 및 누적확률밀도분포 그리기
# scipy.stats 패키지
# norm.cdf() 함수 = 누적밀도 함수
# stats.norm.cdf(x, loc, scale)
# (확률변수값, 평균, 표준편차)
# 즉, pob는 scipy 패키지 내 누적밀도 함수 norm.cdf()로 확률변수값, 평균, 표준편차를 기입한다.

# 소수점 한자리까지 출력
```

```python
# p251 연습
# 이항 분포
# 도장공정에서 광택도 불량 40%나 된다고 함
# 3대의 차량을 임의로 선택했을 때 불량대수가 각각 0, 1, 2, 3대가 나올 확률은 얼마인가요?

# n의 수
n = 3 # 3대 차량
for i in range(n + 1): # 0, 1, 2, 3
    prob6 = stats.binom.pmf(k = i, n = n, p = 0.4) # 0. 3, 40% 불량률
    print("P(X={0}) = {1:.3f}".format(i, prob6))
    
# 차량 3대의 불량대수 0, 1, 2, 3을 기준으로 불량 40%를 체크
# 이항확률질량분포 binom.pmf는 k, n, p로 구성
# k는 전체 시행 가운데 성공 횟수(구하는 확률값)
# n은 전체 시행횟수
# p는 독립시행의 성공확률
```

```python
# p271 연습
# 추론 통계
# 새로운 품종의 배추 40통을 랜덤추출하여 무게를 측정할 결과가 다음과 같음
# (단, 모표준편차는 0.397)
# 이 자료로부터 신품종 배추의 평균 무게에 대한 95% 신뢰구간을 추정.

# 1. 직접 계산 시
# : x_bar = 2.7, 시그마 = 0.397, Z0.05/2 = 1.96, n = 40이므로, 신뢰구간 공식에 대입하면 신품종 배추의 평균무게에 대한 95% 신뢰구간 = (2.57, 2.82)
# → 시그마 0.397 = 모표준편차
# → n 40 = 배추 40통
# → 1.96

# 2. 모평균 신뢰구간 추정
# df를 직접 입력
df = pd.DataFrame({"sample" : [3.6, 2.9, 2.8, 2.6, 3.4, 3.2, 2.2, 2.6, 2.6, 2.4, 2.4, 2.6, 3.4, 2.4, 2.1, 2.6, 3.0, 2.4, 2.7, 2.4, 2.6, 2.9, 2.9, 2.0, 2.7, 2.2, 2.8, 2.7, 1.8, 2.5, 3.0, 3.2, 2.8, 2.6, 3.2, 3.1, 2.9, 2.7, 2.7, 2.2]})
df.head()
```

```python
# 정규분포 구간 추정 함수 실행
lower, upper = stats.norm.interval(0.95, loc = np.mean(df), scale = 0.397/np.sqrt(40))
print("신뢰구간: ({0}, {1})".format(lower.round(2), upper.round(2)))

# Construct a 95% confidence interval
# 표본의 개수가 30개 이상이라면(= 표본의 개수가 충분할 때), stats.norm.interval(alpha, loc, scale)
# alpha는 0.95
# loc는 샘플의 평균
# scale은 std 즉, sqrt 40으로 나눈 값이 std() 가 된다.

# 그리고 lower와 upper을 소숫점 두자리 수까지 표현한다.
```

```python
# p300 연습
# 추론 통계 (2 sample t-test) :, 두 집단 간 평균이 같은지 비교
# 고객만족도 확인을 위하여 A고객과 B고객에 대해 Survey하여 다음의 데이터를 얻었다. 
# A고객과 B고객의 모평균에 차이가 있다고 할 수 있는가? (유의수준 a = 0.05)
# 데이터 (10점 척도, n = 20)

df = pd.DataFrame({'A' : [6, 5, 5, 4, 6, 7, 6, 4, 5, 6, 4, 5, 5, 6, 4, 8, 6, 5, 6, 7], 'B' : [7, 6, 7, 8, 7, 8, 8, 5, 7, 6, 5, 5, 6, 6, 5, 7, 9, 7, 7, 8]})
df  
```

```python
# p319 연습
# 추론 통계 (chi-square test)
# 근무조별로 철강제품을 생산하고 있는데, 생산 제품의 규격이 3가지 종류가 있다. 
# 근무조별로 제품을 생산하는 부하는 차이가 있겠는가? (유의수준 0.05)

df = pd.DataFrame({'Work_1': [270, 228, 277], 'Work_2': [260, 285, 284], 'Work_3': [236, 225, 231], 'Work_4': [234, 262, 208]})
df
```

```python
# p326 연습
# 추론 통계 (ANOVA)
# 회사 마케팅실 그룹별 직원들의 TOEIC 성적에 차이가 있는지를 알고 싶다. 
# A 그룹,  B 그룹, C 그룹 간의 토익 점수의 차이 여부를 검정하세요 (유의수준 a = 0.05)

df = pd.DataFrame({'A': [892, 623, 721, 678, 723, 790, 720, 670, 690, 771], 'B': [721, 821, 910, 678, 723, 790, 711, 790, 745, 891], 'C': [750, 915, 888, 721, 894, 834, 841, 912, 845, 889]})
df
```

```python
# group별 boxplot
df.boxplot(['A', 'B', 'C'])
```

```python
# 'A', 'B', 'C' 정규성 검정 : shapiro-Wilk Test
statistic, p = stats.shapiro(df['A'])
print("statistic={}, p-vaule={}".format(statistic, p))

statistic, p = stats.shapiro(df['B'])
print("statistic={}, p-vaule={}".format(statistic, p))

statistic, p = stats.shapiro(df['C'])
print("statistic={}, p-vaule={}".format(statistic, p))
```

```python
# 'A', 'B', 'C' 등분산성 검정 : bartlett Test
stats.bartlett(df['A'], df['B'], df['C'])
```

```python
# ANOVA 실행
f_result = stats.f_oneway(df['A'], df['B'], df['C'])
f_result
```

```python
# ANOVA의 값 f값 p값 저장
f, p = f_result.statistic.round(3), f_result.statistic.round(3)
print("One-way ANOVA")
print("F 통계량: {}".format(f))
print("p-value: {}".format(p))

# 결론 : 유의수준 5%에서 p_value가 0.003이므로 그룹간 토익점수의 평균은 차이가 있다고 할 수 있다.
```
