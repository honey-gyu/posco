```python
# p36 실습
# 정규분포에서 추출한 10개의 무작위 데이터로 slice, loc, iloc, query 등을 이용하여 데이터를 선택하라.
# 데이터 : 정규분포에서 추출한 10개의 무작위 데이터
# (변수 : 4개, 자료 수 : 10개, random seed = 1234)

# [1] 데이터 구성하기
# [2] slice 1. B열 선택 2. B, D열 선택 3. 4~8행 선택
# [3] loc 1. 2~8행, A,C열 선택 2. A, C열 선택 3. 3~8행 선택 4. 3행 선택(단일행)
# [3-1] ILOC 1. 0~6행, a열 선택 2. A열 선택, 3. 2~6행 선택 4. 5행 선택(단일행)
# [4] query 1. D > 0.5 데이터 선택 2. B > D 데이터 선택 (그리고, slice 표현으로도 데이터 선택) 3. A > 0.5 그리고 D < 0.0 데이터 선택 4. A > 0.5 혹은 D < 0-.0 데이터 선택

import pandas as pd
import numpy as np
from numpy.random import randn

np.random.seed(1234)
rand_data = pd.DataFrame(randn(10, 4), columns = ['A', 'B', 'C', 'D']) # [1]
rand_data
```

```python
# p50 실습
# 아래와 같은 샘플 데이터 df4, df5, df6을 이용해 concat(좌우 결합)을 실행하라.
# 데이터 : 직접 생성한 아래와 같은 df4, df5, df6 데이터 (변수 4개, 자료 수 4개)

# [1] 데이터 구성하기
# [2] concat을 이용한 데이터 결합

df4 = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'],
                   'B': ['B0', 'B1', 'B2', 'B3'],
                   'C': ['C0', 'C1', 'C2', 'C3'],
                   'D': ['D0', 'D1', 'D2', 'D3']},
                   index = [0, 1, 2, 3])
df5 = pd.DataFrame({'E': ['E4', 'E5', 'E6', 'E7'],
                   'F': ['F4', 'F5', 'F6', 'F7'],
                   'G': ['G4', 'G5', 'G6', 'G7'],
                   'H': ['H4', 'H5', 'H6', 'H7']},
                   index = [0, 1, 2, 3])
df6 = pd.DataFrame({'I': ['I8', 'I9', 'I10', 'I11'],
                   'J': ['J8', 'J9', 'J10', 'J11'],
                   'K': ['K8', 'K9', 'K10', 'K11'],
                   'l': ['l8', 'l9', 'l10', 'l11']},
                   index = [0, 1, 2, 3])

result = pd.concat([df4, df5, df6]) #[1]
result
```

```python
result = pd.concat([df4, df5, df6], axis = 1) #[2]
result
```
