```python
# p126 시작
import pandas as pd # 데이터 구성 패키지 : Series, DataFrame
import numpy as np # 행렬 연산 패키지
from sklearn.preprocessing import StandardScaler, MinMaxScaler, RobustScaler # Scaling 클래스
import matplotlib # 그래프 표시
%matplotlib inline
```

```python
df_raw = pd.read_csv("/home/piai/다운로드/2주차 실습파일/1 Big Data 분석 기초/FITNESS_정제.csv", encoding = 'euc-kr')
df_raw
# GENDER와 WEIGHT 변수에 결측치 존재
```

```python
# 결측치 확인 :
df_raw.isnull()
# 1번 4번 18번 결측치 존재
```

```python
df_raw.isna()
# 값이 결측지(NaN)이면 True, 값이 있으면 False로 나타냄
```

```python
# 결측치(True) 현황 요약 : sum 함수
df_raw.isnull().sum()
# GENDER 변수에 1개의 결측치 존재
# WEIGHT 변수에 2개의 결측치 존재
```

```python
# 범주형(문자) 변수 결측치 처리 : GENDER 변수의 결측치에 "여성"으로 대체
df_raw["GENDER"].fillna("여성", inplace = True)
df_raw
# -> GENDER 행에서 True(결측치)라면 "여성"으로 단어를 바꿔라. (Fillna)
```

```python
# 연속형(숫자) 변수 결측치 처리 : GENDER별 WEIGHT 평균 계산 및 대체

# 숫자형 결측치 처리 : groupby(범주형 변수).agg("집계함수")
df_raw.groupby("GENDER")["WEIGHT"].agg("mean")
#-> 성별 간 몸무게 평균

# 18번 라파엘은 몸무게가 NaN이다. 이럴 경우
# "평균"으로 대체한다면 남성의 WEIGHT 평균으로 대체해야 함 (남성과 여성의 평균 몸무게는 차이 고려)
# -> 다른 대체 방법 적용 가능 : 중앙값 (median) 등
```

```python
# 연속형(숫자) 변수 결측치 처리 : WEIGHT 변수의 결측치 처리
df_raw["WEIGHT"] = df_raw["WEIGHT"].fillna(df_raw.groupby("GENDER")["WEIGHT"].transform("mean")).round(3)
df_raw 
# WEIGHT 변수의 결측치를 GENDER = "남성"의 WEIGHT 평균으로 대체
# -> 몸무게 행에 fillna(NaN)이 있다면 채워라. 남성의 평균값을 mean으로 변경해서. 그리고 소숫좀 3자리 수까지
```

```python
# 이상치 확인 및 처리 : 상자수염그래프를 이용한 이상치 확인

# 단위와 같이 표현되어 있는 수치형 변수 처리
df_raw[['RUNTIME', 'RUNTIME_UNIT']] = df_raw['RUNTIME'].str.split(' ', expand = True)
df_raw['RUNTIME'] = df_raw['RUNTIME'].astype('float64')

# boxplot
df_raw.boxplot(figsize = (10, 4))

# 왜 한꺼번에 있어야만 실행이 될까 
# 왜 또 실행하면 오류가 날까
```
