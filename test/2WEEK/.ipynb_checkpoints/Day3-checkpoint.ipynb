```python
df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'],
                   'B': ['B0', 'B1', 'B2', 'B3'],
                   'C': ['C0', 'C1', 'C2', 'C3'],
                   'D': ['D0', 'D1', 'D2', 'D3']},
                   index = [0, 1, 2, 3])
df2 = pd.DataFrame({'A': ['A4', 'A5', 'A6', 'A7'],
                   'B': ['B4', 'B5', 'B6', 'B7'],
                   'C': ['C4', 'C5', 'C6', 'C7'],
                   'D': ['D4', 'D5', 'D6', 'D7']},
                   index = [4,5,6,7])
df3 = pd.DataFrame({'A': ['A8', 'A9', 'A10', 'A11'],
                   'B': ['B8', 'B9', 'B10', 'B11'],
                   'C': ['C8', 'C9', 'C10', 'C11'],
                   'D': ['D8', 'D9', 'D10', 'D11']},
                   index = [8,9,10,11])

result = pd.concat([df1, df2, df3])
result
```

```python
result = pd.concat([df1, df2, df3], axis = 1)
result
```

```python
result = pd.concat([df1, df2, df3], axis = 1, ignore_index=True)
result
```

```python
# 오후 수업 시작
# filter, mask
import pandas as pd
choice_data = {"ABC" : ['A', 'B', 'C', 'A'],
              "abc" : ['a', 'b', 'c', 'a'],
              "group1" : ['G1', 'G2', 'G3', 'G4'],
              "value1" : [11, 13, 12, 14],
              "val1_mod" : [0.231, 0.197, 0.213, 0.199]}
table = pd.DataFrame(choice_data)
table
```

```python
# filter 함수 
# items는 입력한 컬럼만 필러해서 출력해줘
# table.filter(items = None, like = None, regex = None, axis = None)
table.filter(items=['group1', 'value1'])
# 컬럼이름 직접 입력
```

```python
table.filter(like = 'a', axis = 1)
# axis = 1이 뭘까?
# 'a'가 포함된 컬럼. regex='a'와 동일
```

```python
table.filter(regex = '1', axis = 1)
```
